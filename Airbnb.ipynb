Here, preprocessing is done to make the columns usable. The dollar sign is removed and the column is converted to numeric.The code for this part is taken from https://www.kaggle.com/qclodd/berlin-airbnb-analysis as well as the functions to extract specific facilities from amenities. For bedroom, bathroom and beds, the mean value is imputed for rows with missing values. 

```python
#preprocessing the data 

df_listing["security_deposit"].fillna("$0.00", inplace=True)
df_listing["cleaning_fee"].fillna("$0.00", inplace=True)

# Removing $ sign from price columns - some numbers contain "," in them - gonna delete that comma there
def remove_dollar(str_price):
    if "," in str_price:
        str_price = str_price.replace(",", "")
    else:
        pass
    return str_price[1:]


for dollar_column in ["price", "security_deposit", "cleaning_fee"]:
    price_no_dollar = df_listing[dollar_column].apply(remove_dollar)
    df_listing[dollar_column] = price_no_dollar
    
# Converting columns to "float64" dtype
price_num = pd.to_numeric(df_listing["price"])
security_deposit_num = pd.to_numeric(df_listing["security_deposit"])
cleaning_fee_num = pd.to_numeric(df_listing["cleaning_fee"])


# Replacing the old columns
df_listing["price"] = price_num
df_listing["security_deposit"] = security_deposit_num
df_listing["cleaning_fee"] = cleaning_fee_num


```

```python
# find mean for bedroom, beds, and bathroom and fill the remaining with that. 
print(df_listing["bedrooms"].mean())
print(df_listing["bathrooms"].mean())
print(df_listing["beds"].mean())

df_listing["bedrooms"].fillna(1.0, inplace=True)
df_listing["bathrooms"].fillna(1.0, inplace=True)
df_listing["beds"].fillna(2.0, inplace=True)
```

```python
df_listing = df_listing.dropna(axis=0, subset=['host_is_superhost'])

df_listing.isnull().sum()
```

```python
# putting average reviews so that I can make the score.the remaining ones that were not dropped, we just imputed the mean. 

print(df_listing["review_scores_location"].mean())
print(df_listing["review_scores_accuracy"].mean())
print(df_listing["review_scores_cleanliness"].mean())
print(df_listing["review_scores_checkin"].mean())
print(df_listing["review_scores_communication"].mean())
print(df_listing["review_scores_value"].mean())

df_listing["review_scores_location"].fillna(10, inplace=True)
df_listing["review_scores_accuracy"].fillna(10, inplace=True)
df_listing["review_scores_cleanliness"].fillna(9, inplace=True)
df_listing["review_scores_communication"].fillna(10, inplace=True)
df_listing["review_scores_checkin"].fillna(10, inplace=True)
df_listing["review_scores_value"].fillna(9, inplace=True)
```

The reviews for most listings are high. Therefore, it is not a good measure to rank a listing. Although I did preprocess the review_scores, I won't be using it any further. 

The true/false values are converted to binary 1/0 respectively.

```python
#converting to workable columns 
def conversion(value):
    if value == "t":
        return 1
    else:
        return 0
    
df_listing["host_is_superhost"] = df_listing["host_is_superhost"].apply(conversion)
df_listing["require_guest_profile_picture"] = df_listing["require_guest_profile_picture"].apply(conversion)
df_listing["require_guest_phone_verification"] = df_listing["require_guest_phone_verification"].apply(conversion)
df_listing["instant_bookable"] = df_listing["instant_bookable"].apply(conversion)
```

```python
#feature generation

#we can extract important features from amenities such as TV, WIFI,heating.

def wifi(value):
    if "wifi" in value.lower():
        return 1
    else:
        return 0

def tv(value):
    if "tv" in value.lower():
        return 1
    else:
        return 0

def heating(value):
    if "tv" in value.lower():
        return 1
    else:
        return 0

def wheelchair(value):
    if "tv" in value.lower():
        return 1
    else:
        return 0
    
df_listing["Wifi"] = df_listing["amenities"].apply(wifi)
df_listing["Tv"] = df_listing["amenities"].apply(tv)
df_listing["Heating"] = df_listing["amenities"].apply(heating)
df_listing["Wheelchair accessible"] = df_listing["amenities"].apply(wheelchair)

```

Amenities is dropped as we won't be needing it further. Here, the columns that have been preprocessed are also dropped. I also dropped the columns that I will not be needing for ranking. eg. has_availability is dropped as all the values are the same. Some of the values have no direct correlation to price or facilities,and hence they are dropped as well. Latitude and Longitude is also dropped because the neighbourhood_group_cleansed is already used for location. Therefore, we won't be needing it further. 

```python
df_listing.drop(["neighbourhood_group_cleansed","review_scores_location","review_scores_accuracy","review_scores_cleanliness","review_scores_communication",
"review_scores_checkin", "review_scores_value","review_scores_rating","latitude","longitude"], axis=1, inplace=True)
df_listing.drop(["amenities", "has_availability","bed_type","cancellation_policy","room_type","last_scraped","host_name","host_since","calendar_updated", "availability_30","availability_60","availability_90"], axis=1, inplace=True)
```

## Scoring 

The important amenities are added together to give a relative amount (score) of facilities given by the place. Similarly, the column giving various services are added to give a service score. Each of them are converted to a rating of 5. There are five categories in my ranking system- facilities, property, accomodates,service and location. Using these individual ranks, wecan come up with a holistic idea of the listing. 


```python
df_listing["facilities_score"]= ((df_listing["Wifi"]+ df_listing["Tv"]+ df_listing["Heating"]+ df_listing["Wheelchair accessible"])/4)*5
df_listing["property_score"]= ((df_listing["bed_type_rank"]+ df_listing["room_type_rank"])/6)*5
df_listing["accomodates_score"]=(df_listing["accommodates"]/16)*5
df_listing["service_score"]= ((df_listing["cancellation_rank"]+ df_listing["instant_bookable"]+df_listing["require_guest_profile_picture"]+ 
    df_listing["require_guest_phone_verification"]+df_listing["host_is_superhost"])/8)*5
df_listing["location_score"]=(df_listing["neighbourhood_rank"]/7)*5
```

```python
df_listing["score"]= df_listing["facilities_score"]+df_listing["location_score"]+df_listing["property_score"]+df_listing["accomodates_score"]+df_listing["service_score"]
```

```python
df_listing.corr()
```

Noting the correlation, it can be seen that location is a very important factor in the scoring. This has been proved by a lot of data visualisation as well. Also, the score does have a higher correlation with price than other factors. From that we can also deduce that there is no singular factor that determines the price of the listing.

# Comparing listings in Paris and London to that of Berlin

One way to check whether London is 20% more expensive than France would be to check the prices of similar property types at both the places. We do not have data for all of France, and therefore, we will be using Paris as a representation of France. I will be using the columns property_type and price to check the statement of the question. Other factors such as cleaning fee could be taken into account, but housing would definitely be the most important factor. Therefore, we could just focus on that. The weekly price or monthly price could also be used to get an estimate, but they had null values, and so, price would be the best measure. 

First the dataset are uploaded, and the necessary columns are checked for missing values. The price column is preprocessed to remove the $ sign.  

```python
listing_london = pd.read_csv('London.csv')
listing_paris = pd.read_csv('Paris.csv')
```

```python
print(listing_paris["price"].isnull().sum())
print(listing_paris["property_type"].isnull().sum())

print(listing_london["price"].isnull().sum())
print(listing_london["property_type"].isnull().sum())
```

For paris, plotting median price in y-axis and property_type in x-axis,
![alt text](paris.png "paris")

For london, plotting median price in y-axis and property_type in x-axis,
![alt text](london.png "london")




```python
for dollar_column in ["price"]:
    price_no_dollar = listing_london[dollar_column].apply(remove_dollar)
    listing_london[dollar_column] = price_no_dollar
       
# Converting columns to "float64" dtype
price_num = pd.to_numeric(listing_london["price"])
listing_london["price"] = price_num

for dollar_column in ["price"]:
    price_no_dollar = listing_paris[dollar_column].apply(remove_dollar)
    listing_paris[dollar_column] = price_no_dollar

price_num = pd.to_numeric(listing_paris["price"])
listing_paris["price"] = price_num
```

Then the unique property_types are taken and they are sorted according to alphabetical order. This is done so that later on, joining the data with its respective price is easier. The median prices for each of the property_types are found and two dataframes are made- one for london and another for paris. Each dataframe consists of property_type and its respective median price. It should be mentioned here that median price is taken so that the values are not affected by outliers. 

```python
p_paris= sorted(listing_paris.property_type.unique())
p_london= sorted(listing_london.property_type.unique())

print(p_paris)
print(p_london)

median_paris = listing_paris.groupby("property_type")["price"].median()
median_london = listing_london.groupby("property_type")["price"].median()

print(median_london)
print(median_paris)

paris_medianprice = median_paris.tolist() 
london_medianprice = median_london.tolist() 
```

```python
df_london=pd.DataFrame(list(zip(p_london, london_medianprice)),
              columns=['property','price_l'])
df_paris=pd.DataFrame(list(zip(p_paris, paris_medianprice)),
              columns=['property','price_p'])

print(df_london)
print(df_paris)
```

The dataframes are then merged into a single dataframe called df_merge. df_merge only consists of those property_types that are common to both cities so that we can compare their prices. Next, I calculated a percentage difference in price with respect to price in Paris to see if the assumption given is true. The hypothesis here is that if more than 50% of the properties have a value higher than 20%, then we can say that London is at least 20% more expensive. 

```python
df_merge = pd.merge(df_london, df_paris)
df_merge["percentage"]= ((df_merge["price_l"]-df_merge["price_p"])/df_merge["price_p"])*100
print(df_merge)
```

According to our results, however, the hypothesis is proved incorrect. Most of the property in Paris is a lot more expensive than in London.This is shown by the fact that most of the percentages were negative. 

It could be that Paris is not a good sample of the entire France,and therefore it is not e good representation of France. Just because PARIS is more expensive than London does not mean that will be the case if the entire France is taken into account. Perhaps, the others areas are much cheaper and that evens out the high price of Paris.
